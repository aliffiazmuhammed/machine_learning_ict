```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('/content/insurance.csv', index_col=0)

# display the first 5 rows
data.head(10)
```

```python
col = ['sex','region','smoker']
data = data.drop(col,axis =1)
```

```python
data['New_Index'] = range(len(data))
data.head()
```

```python
data.plot(kind='scatter',x='bmi',y='charges',color='red')
data.plot(kind='scatter',x='children',y='charges',color='blue')
data.plot(kind='scatter',x='age',y='charges',color='green')
```

```python

x = data[['bmi','children','age']]
y =  data['charges']
```

```python
x.head(3)
```

# **LINEAR REGRESSION**

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=1)
```

```python
from sklearn.linear_model import LinearRegression


linreg = LinearRegression()


model = linreg.fit(x_train, y_train)
```

```python
y_pred = linreg.predict(x_test)
print(y_pred)
```
